## Métodos
### meanDay()
Calcula y regresa un pandas.DataFrame con la Temperaturas e Irradiancias del día promedio
###### Puede Modificar el día, mes y año a reportar pasandoselos como atributos

###### Si no se especifica día, mes y año, se utilizan los valores predeterminados,
###### day = "15", month = "current" y year = "current" para usar el mes y año del equipo

```python
temixco.meanDay().info()
```

```python
# Acepta enteros o string's, NO acepta meses por nombre (ej. "July")
temixco.meanDay(month= "6", year = 2020).info()
```

### flag()
##### Use flag() para saber si el método meanDay() recalculó el modelo o consultó su valor guardado en la última ejecución
Regresa una copia del ultimo estado del diccionario de la bandera interna

```python
temixco.flag()
```

```python
bandera = temixco.flag()

rec = bandera["recalculate"]
fecha_1 = bandera["date"]
fecha_2 = bandera["day"] +"/"+ bandera["month"] +"/"+ bandera["year"]

print(f"Se recalculó? {rec}")
print(f"Fecha de bandera: {fecha_1}")
print(f"Fecha construida: {fecha_2}\n")
```
