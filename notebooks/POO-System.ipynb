# Libreta para mostrar los métodos y atributos de la clase System
Los atributos se inician con sus valores predeterminados si no se definen al inicializar la instancia

El valor predeterminado de atributo layers es una lista vacia pero deben definirse las capas antes de usar los métodos solve() y solveAC()

##### El atributo location es requerido por lo que la clase System debe inicializarse pasandole una clase Location

```python
import enerhabitat as eh
eh.config.file = "../eh_config/materials.ini"

epw_file = "../epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2004-2018.epw"

muro = eh.System(eh.Location(epw_file))
muro.info()
print()
muro.location.info()
```

### Los atributos pueden ser leidos y sobreescritos

```python
muro.tilt = 0
muro.azimuth = 45
muro.absortance = 0.3

muro.info()
```

### layers
El atributo layers define las capas del sistema constructivo de afuera hacia adentro, es una lista de tuplas de tipo (Material, Espesor)
ej. [('Brick',0.1), ('Insulation',0.05), ('Adobe',0.02)]

Puede sobreescribirse directamente o usando los métodos 
#### add_layer(Material, Espesor) 
#### remove_layer(índice_de_capa)

```python
muro.layers = [('Adobe',0.1), ('Acero',0.05), ('Ladrillo',0.02)]
print(muro.layers)

muro.add_layer("Mortero", 0.2)
print(muro.layers)

muro.remove_layer(2)
print(muro.layers)
```

## Métodos
### Tsa()

Calcula y regresa el DataFrame con la Tsa e Is en base al resultado del método meanDay() del atributo location

```python
muro.Tsa().info()
```

##### Para obtener la Tsa estimada de una fecha especifica modifiquelo desde el atributo location

```python
muro.location.meanDay(month=6, year=2025)
muro.info()
```

```python
muro.Tsa().info()
```

### solve() y solveAC()

Regresa la Serie con la Ti estimada con los valores de temperatura y fechas de Tsa()

##### !!! Asegurese de haber definido las capas en "System.layers" de acuerdo a las etiquetas definidas en el archivo eh.config.file

```python
muro.solve()
muro.info()
```

##### La energia total transmitida asi como la de enfriamiento y calentamiento para el caso de solveAC pueden ser accedidas a través de sus atributos

```python
data = muro.solve()

print(f"Energia transferida: {muro.energy_transfer}\n")
data.info()
```

#### Use el método solveAC() para resolver la energia de calentamiento y enfriamiento de un sistema con aire acondicionado y Ti constante

```python
data = muro.solveAC()

print(f"Energia de enfriamiento: {muro.cooling_energy}")
print(f"Energia de calentamiento: {muro.heating_energy}\n")
data
```
